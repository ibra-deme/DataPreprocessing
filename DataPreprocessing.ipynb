# Importer les **librairies**


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Importer le **dataset**

```python
dataset = pd.read_csv("DataPreprocessing.csv")
dataset.head(4)
```

```python
X=dataset.drop(columns=['Online Shopper'])
X
```

```python
Y=dataset['Online Shopper']
Y
```

# gestion des donnees **manquantes**

```python
X
```

```python
X['Income'].mean()
```

```python
#importer imputer depuis sklearn
from sklearn.impute import SimpleImputer
#creer une instance de la classe
imputer=SimpleImputer(missing_values=np.nan, strategy='mean')
#ajuster avec fit
imputer.fit(X[["Age","Income"]])
#transformer
X[["Age","Income"]]=imputer.transform(X[["Age","Income"]])
```

```python
X.isna().sum()
```

```python
X.iloc[: , 1:3]
```

# Encoder les donnees **categorielle**

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers = [('encoder', OneHotEncoder(), [0])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
print(X[0])
```

# Encoder la variable dependante

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
Y = le.fit_transform(Y)
```

```python
print(Y)
```

# **Diviser le dataset en train et test**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=1)
```
